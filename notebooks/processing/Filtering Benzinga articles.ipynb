```python
import json
import pandas as pd

from IPython.display import display
```

```python
# Read Benzinga articles dataset
file = "../../data/raw/raw_analyst_ratings.csv"
articles = pd.read_csv(file, index_col=0)

display(articles.head())
print(len(articles))
```

```python
# Read latest stock metadata dataset
file = "../../data/processed/available_stocks/revolut.top50.2021-08-19T17:50:12.jsonl"

stocks = list()
with open(file, "r") as fp:
    for line in fp:
        stocks.append(json.loads(line))
stocks = [s for s in stocks if s["ticker"] != "TIGR"]
stocks = pd.DataFrame(stocks)

display(stocks.head())
print(len(stocks))
```

```python
# Stocks that appear in articles:
appears_in_articles = set(articles[articles.stock.isin(stocks.ticker)].stock)

display(stocks[stocks.ticker.isin(appears_in_articles)])
```

```python
# Stocks that DON'T appear in articles:

does_not_appear = set(stocks.ticker) - appears_in_articles
display(stocks[stocks.ticker.isin(does_not_appear)])
```

```python
stocks[stocks.ticker.isin(appears_in_articles)][["name", "ticker"]]
```

```python
# Sampling the articles

all_samples = []

for ticker in appears_in_articles:
    print(ticker)
    sampled_articles = articles[articles.stock == ticker].sample(5)
    display(sampled_articles)
    
    all_samples += sampled_articles.to_dict(orient="records")
    
print(len(all_samples))
```

```python
path = "../../data/processed/articles_test/"
file = "benzinga_samples.jsonl"

with open(path + file, "w") as fp:
    for sample in all_samples:
        fp.write(json.dumps(sample))
        fp.write("\n")
```
