```python
import wikipedia
import re
```

```python
wiki_test_pages = {
    'Amazon': 'Amazon_(company)'
}
```

```python
page = wikipedia.page(title=wiki_test_pages['Amazon'])
```

```python
print(page.content)
```

```python
def valid_paragraph(paragraph):
    return len(paragraph) > 0 and not paragraph.isspace() and not paragraph.startswith('==')
 
plaintext = '\n'.join([paragraph for paragraph in page.content.split('\n') if valid_paragraph(paragraph)])
```

```python
print(plaintext)
```
