```python
a = np.array([[1,1,1,1,0,0,0],[1,0,1,0,2,2,3], [1,2,4,5,6,7,4]])
b = np.array([[1,1,1,1,0,0,0],[1,0,1,0,2,1,2], [2,1,4,5,3,5,3]])

np.einsum('ij,ij->i', a, b) / (np.linalg.norm(a, axis=1)*np.linalg.norm(b, axis=1))
```
